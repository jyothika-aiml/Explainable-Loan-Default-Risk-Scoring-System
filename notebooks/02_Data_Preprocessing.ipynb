# Explainable Loan Default Risk Scoring System

# Notebook 2: Data Preprocessing

In this notebook, the loan default dataset is preprocessed before training machine learning models.

The preprocessing steps include:

- Loading the dataset
- Inspecting the data
- Removing unnecessary columns
- Encoding categorical variables
- Splitting the dataset into training and testing sets
- Saving the trained label encoders for deployment

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

import joblib
import os
```

# Load the Dataset

The loan default dataset is loaded into a Pandas DataFrame.

```python
df = pd.read_csv("../data/Loan_Default.csv")

df.head()
```

# Inspect the Dataset

Understanding the structure of the dataset before preprocessing.

```python
print("Shape of Dataset:", df.shape)

df.info()
```

```python
df.isnull().sum()
```

# Remove Unnecessary Columns

The LoanID column is only an identifier and does not contribute to prediction.

Therefore, it is removed before model training.

```python
df.drop("LoanID", axis=1, inplace=True)

df.head()
```

# Encode Categorical Features

Machine learning algorithms require numerical input.

Each categorical feature is encoded using a separate LabelEncoder.

The trained encoders are stored for later use during deployment to ensure consistent preprocessing.

```python
label_encoders = {}

categorical_columns = df.select_dtypes(include=["object", "string"]).columns

print("Categorical Columns:")
print(categorical_columns)
```

```python
for col in categorical_columns:

    encoder = LabelEncoder()

    df[col] = encoder.fit_transform(df[col])

    label_encoders[col] = encoder

df.head()
```

# Verify Encoded Dataset

The categorical columns have now been converted into numerical values suitable for machine learning.

```python
df.head()
```

# Split Features and Target

The dataset is divided into:

- Features (X)
- Target Variable (y)

```python
X = df.drop("Default", axis=1)

y = df["Default"]

print("Feature Shape :", X.shape)
print("Target Shape  :", y.shape)
```

# Train-Test Split

The dataset is divided into training and testing sets.

- 80% Training Data
- 20% Testing Data

Stratified sampling is used to preserve the class distribution.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Shape :", X_train.shape)
print("Testing Shape  :", X_test.shape)
```

# Save Label Encoders

The trained LabelEncoders are saved for deployment.

This ensures that the Streamlit application uses the exact same categorical encoding that was used during model training.

```python
os.makedirs("../models", exist_ok=True)

joblib.dump(label_encoders, "../models/label_encoders.pkl")

print("Label encoders saved successfully!")
```

# Save Feature Names

The feature names are saved so that the deployment application and SHAP explanations use the correct feature order.

```python
joblib.dump(list(X.columns), "../models/feature_names.pkl")

print("Feature names saved successfully!")
```

# Verify Saved Files

The saved preprocessing artifacts are loaded to ensure they were stored successfully.

```python
encoders = joblib.load("../models/label_encoders.pkl")

print("Saved Encoders:")
print(encoders.keys())
```

```python
feature_names = joblib.load("../models/feature_names.pkl")

print(feature_names)
```

# Conclusion

The dataset has been successfully preprocessed for machine learning.

The preprocessing pipeline included:

- Removing unnecessary columns
- Encoding categorical variables
- Splitting the dataset into training and testing sets
- Saving LabelEncoders for deployment
- Saving feature names for consistent model inference

The processed data is now ready for model training in the next notebook.
